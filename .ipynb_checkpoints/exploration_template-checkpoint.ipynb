> This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

> The data set is filtered to affecting variables only and records with NAN/NA values are deleted.

### Project Output

In this explanatory study, we tend to study and address the following questions:
> What factors affect a loan’s outcome status?

> What affects the borrower’s APR or interest rate?

> Are there differences between loans depending on how large the original loan amount was?

```python
# Get the unique values of Loan Status
df.LoanStatus.unique()
# Past Due values have another attribute attached to it.
```

```python
#Split the LoanStatus Column to get only the Status
df[['LoanStatus2','LoanStatusPastDue']]= df.LoanStatus.str.split('(',expand=True,);
```

```python
#Remove ')' from the last column LoanStatusPastDue
df['LoanStatusPastDue'] = df['LoanStatusPastDue'].str.replace(')', '')
```

```python
# map category names instead of numeric categories
df['ListingCategory'] = df['ListingCategory (numeric)'].map({0: 'Not Available', 1: 'Debt Consolidation', 2 : 'Home Improvement', 3: 'Business' , 4 : 'Personal Loan' , 5 : 'Student Use' , 6 : 'Auto' , 7 : 'Other' , 8 : 'Baby&Adoption' , 9 : 'Boat' , 10 : 'Cosmetic Procedure' , 11 : 'Engagement Ring' , 12 : 'Green Loans' , 13 : 'Household Expenses' , 14 : 'Large Purchases' , 15 : 'Medical/Dental' , 16 : 'Motorcycle' , 17 : 'RV' , 18 : 'Taxes' , 19 : 'Vacation' , 20 : 'Wedding Loans'}).fillna(df['ListingCategory (numeric)'])
```

```python
# Convert days to years
def get_years_from_days(x, inverse = False):
    if x > 0:
        if not inverse:
            return x/360
        else:
            return x*360
        
# Convert Months to years
def get_years_from_months(x, inverse = False):
    if x > 0:
        if not inverse:
            return x/12
        else:
            return x*12
        
# Convert values to log10        
def apply_log10(x, inverse = False):
    if not inverse:
        return math.log(x,10)
    else:
        return 10**x
    
# Convert Values to log
def apply_log(x, inverse = False):
    if not inverse:
        return np.log(x)
    else:
        return np.exp(x)

#Convert values to 1/x
def div_inverse(x):
    if x == 0:
        return 0
    else:
        return 1/x
```

```python
#Plot Monthly Income Ration to Monthly Income
plt.figure(figsize = [12,5])
plt.scatter(x = df.DebtToIncomeRatio.apply(div_inverse), y = df.StatedMonthlyIncome, alpha = 0.2);
plt.ylim(0,120000);
plt.xlabel('Loan to Income Ratio');
plt.ylabel('Monthly Income ($)');
plt.title('Monthly Income Ratio vs. Monthly Income');
```

> Borrowers tend to borrow loans, in most cases, up to 20 times greater than their income and in rare cases, up to 100 times greater than their income.

>Yet the majority of Loan borrowers falls in a lower range (between 0 to 10 times more than their income)

> To better understand what forces borrowers from not paying loans back. we need to filter only The number of days delinquent that is more than a year.

```python
# Get dataset of late payments
df_late = df[df['LoanCurrentDaysDelinquent'].apply(get_years_from_days)>1]
```

```python
# #Explore all numeric variables
# # Do not run this code ... It takes 2 hpurs to finish

# numeric_columns = df_late.select_dtypes(include=['float64']).columns.tolist()
# g = sb.PairGrid(data = df_late, vars = numeric_columns)
# g.map_diag(plt.hist)
# g.map_offdiag(plt.scatter)
```

```python
df_late['MonthlyLoanPaymentDividedByMonthlyIncome'] = df_late['MonthlyLoanPayment'] / df_late['StatedMonthlyIncome']
```

```python
# Describe late payments Dataset
df_late['MonthlyLoanPaymentDividedByMonthlyIncome'].describe()
```

```python
# Get only dataset of non zero values of Monthly Loan Payment and category of not (NOt Available)
temp_df = df_late[(df_late.MonthlyLoanPaymentDividedByMonthlyIncome > 0) &
                  (df_late.ListingCategory != 'Not Available')]
```

### Observed Relationships

> There is no direct relationship between the features

### Interesting Relationships

> Borrower Interest rate is directly correlated to the following:
<lu>
    <li> Lend Yield
    <li> Estimated Effective Yield
    <li> Estimated Loss
    <li> Estimated Return
</lu>

```python
# Plot Borrower Interest rate to Lend Yield
plt.figure(figsize=[10,7])
plt.subplot(2,2,1)
plt.scatter(x = temp_df.BorrowerRate, y= temp_df.LenderYield)
plt.xlabel('Borrower Rate')
plt.ylabel('Lender Yield')
plt.title('Borrow Rate vs. Lender Yield')

# Plot Borrower Interest rate to Estimated Effective Yield
plt.subplot(2,2,2)
plt.scatter(x = temp_df.BorrowerRate, y= temp_df.EstimatedEffectiveYield)
plt.xlabel('Borrower Rate')
plt.ylabel('Estimated Effective Yield')
plt.title('Borrow Rate vs. Estimated Effective Yield')

# Plot Borrower Interest rate to Estimated Loss
plt.subplot(2,2,3)
plt.scatter(x = temp_df.BorrowerRate, y= temp_df.EstimatedLoss)
plt.xlabel('Borrower Rate')
plt.ylabel('Estimated Loss')
plt.title('Borrow Rate vs. Estimated Loss')

# Plot Borrower Interest rate to Estimated Return
plt.subplot(2,2,4)
plt.scatter(x = temp_df.BorrowerRate, y= temp_df.EstimatedReturn)
plt.xlabel('Borrower Rate')
plt.ylabel('Estimated Return')
plt.title('Borrow Rate vs. Estimated Return')

plt.suptitle('Borrow Rate Interaction')

plt.tight_layout()
```

## Multivariate Exploration
